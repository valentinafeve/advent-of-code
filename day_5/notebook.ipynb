```python
import string
```

```python
input = open("input", "r")
print(input.read())
```

```python
stacks = {}
lines = open("input", "r").readlines()
for line in lines:
    if not "[" in line:
        break
    cont = 0
    for char in line:
        if char in string.ascii_uppercase:
            idx = str(int((cont+3)/4))
            stacks[idx] = stacks.get(idx, [])
            stacks[idx].insert(0, char)
            print(f"idx: {idx} letter: {char}")
        cont += 1
```

```python
stacks
```

```python
for line in lines:
    if "move" in line:
        line = line.replace("move ", "").replace("from ", "").replace("to ", "").strip()
        n, from_i, to_i = line.split(" ")
        print(f"move {n} from {from_i} to {to_i}")
        stacks[to_i].extend(stacks[from_i][-int(n):])
        [stacks[from_i].pop() for i in range(int(n))]
```

```python
s = ""
for i in range(len(stacks.keys())):
    s += stacks[str(i+1)][-1]
```
